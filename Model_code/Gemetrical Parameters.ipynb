# Relationship btw Heliocentric Distance of ICME apex R$_0$ and element located at angle $\alpha$ R$_\alpha$ for angular width of $\omega$.





```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
def R_alpha(omega):
    omega=np.deg2rad(omega)
    alpha=np.linspace(0,omega)
    R0=1
    ralpha= R0 *(np.cos(alpha) +((np.tan(omega))**2 - (np.sin(alpha))**2)**0.5)/(1+ np.tan(omega))
    NT=[np.rad2deg(alpha),ralpha]
    
    return(NT)
 
    
    
"""
This same function can be use for Velocity profile.
Just change R_alpha to V_alpha (ralpha to vaplha) & R0 to V0.
So we get a same graph for velocity profile as shown below. 
"""
```

```python
plt.figure(figsize=(15,10))
plt.plot(R_alpha(1)[0],R_alpha(1)[1],label="$\omega$ = 1")
plt.plot(R_alpha(5)[0],R_alpha(5)[1],label="$\omega$ = 5")
plt.plot(R_alpha(10)[0],R_alpha(10)[1],label="$\omega$ = 10")
plt.plot(R_alpha(20)[0],R_alpha(20)[1],label="$\omega$ = 20")
plt.plot(R_alpha(30)[0],R_alpha(30)[1],label="$\omega$ = 30")
plt.plot(R_alpha(40)[0],R_alpha(40)[1],label="$\omega$= 40")
plt.plot(R_alpha(50)[0],R_alpha(50)[1],label="$\omega$ = 50")
plt.plot(R_alpha(60)[0],R_alpha(60)[1],label="$\omega$ = 60")
plt.plot(R_alpha(70)[0],R_alpha(70)[1],label="$\omega$ = 70")
plt.plot(R_alpha(80)[0],R_alpha(80)[1],label="$\omega$= 80")
plt.plot(R_alpha(90)[0],R_alpha(90)[1],label="$\omega$ = 90")
plt.legend(loc="lower left",fontsize=15)
plt.xlabel(r"$\alpha$(deg)",fontsize=20)
plt.ylabel(r"$R_\alpha$ / $R_0$",fontsize=20)
plt.grid()
plt.show()
```
